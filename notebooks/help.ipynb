```python
import os
from docx import Document


directory = os.path.join('docx_docs', 'patents')
docs = {}
for file in os.listdir(directory):
    # print(os.getcwd())
    doc = Document(os.path.join(os.getcwd(), directory, file))
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    docs[file] = ' '.join(fullText)
```

```python
import re
import math

df['country'] = "United States of America"
for index, row in merged_data.iterrows():
    new_row = pd.DataFrame({'title': [row['назив проналаска']],
               'register_num': [row['број \nрегистро-\nваног \nпатента']],
               'type': ['patent'],
               'country': [row['земља']],
               'date': [row['датум \nрегистрације \nпатента']],
               'register_num': [row['број \nрегистро-\nваног \nпатента']],
               })
    df = pd.concat([df, new_row], ignore_index=True)
m = {'United States of America': "United States of America", 'Аргентина': "Argentina", 'Аустралија': "Australia", 'Аустрија': "Austria",
       'Белгија': "Belgium", 'Бразил': "Brazil", 'Велика \nБританија': "Great Britain", 'Викторија': "Victoria", 'Данска': "Denmark",
       'Индија': "India", 'Италија': "Italy", 'Јапан': "Japan", 'Канада': "Canada", 'Куба': "Cuba", 'Мађарска': "Hungary",
       'Мексико': "Mexico", 'Немачка': "Germany", 'Нови \nЗеланд': "New Zealand", 'Нови Јужни \nВелс': "New South Vales",
       'Норвешка': "Norway", 'Родезија': "Rhodesia", 'Русија': "Russia", 'Трансвал': "Transvaal", 'Француска': "France",
       'Швајцарска': "Switzerland", 'Шведска': "Sweden", 'Шпанија': "Spain"}
for k, v in m.items():
    df.loc[df['country'] == k, 'country'] = v
df.dropna(subset=['date'], inplace=True)

for index, row in df.iterrows():
    df.at[index, 'date'] = pd.to_datetime(re.sub(r'\([^)]*\)', '', row['date']).replace('. ', '.'))
df['date'] = df['date'].astype(str)
i = 0
for index, row in df.iterrows():
    if math.isnan(row['id']):
        df.at[index, 'id'] = i
        i += 1
    else:
        i = row['id']
```

```python
import sqlite3
# Establish a connection to the database
conn = sqlite3.connect('metadata_new.db')

# Save the DataFrame as a table in the database
df.to_sql('metadata', conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```
